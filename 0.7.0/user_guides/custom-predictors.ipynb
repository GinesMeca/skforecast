# Custom predictors for recursive multi-step forecasting

When forecasting, it may be interesting to include additional characteristics of the time series, besides the lags. For example, the moving average of the last n values can be used to capture the trend of the series.

`ForecasterAutoregCustom` (similar to the `ForecasterAutoreg`, [Recursive multi-step forecasting](https://joaquinamatrodrigo.github.io/skforecast/latest/user_guides/autoregresive-forecaster.html)) and `ForecasterAutoregMultiSeriesCustom` (similar to the `ForecasterAutoregMultiSeries`, [Multi-series forecasting](https://joaquinamatrodrigo.github.io/skforecast/latest/user_guides/multi-time-series-forecasting.html)) cover the case where the user can define the function used to create the predictors.

To do this, the user has to create a function that receives a time series as input (numpy ndarray) and returns another numpy ndarray with the predictors. The Forecaster parameters used to include this function in the class are `fun_predictors` and `window_size`.

The <code>window_size</code> parameter represents the size of the data (window) that <code>fun_predictors</code> needs to create each row of predictors. It is important to give the <b>correct value</b> to this argument in order not to lose data when creating the training matrices.

## Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skforecast.ForecasterAutoregCustom import ForecasterAutoregCustom
from skforecast.ForecasterAutoregMultiSeriesCustom import ForecasterAutoregMultiSeriesCustom
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

## ForecasterAutoregCustom

```python
# Download data
# ==============================================================================
url = ('https://raw.githubusercontent.com/JoaquinAmatRodrigo/skforecast/master/data/h2o.csv')
data = pd.read_csv(url, sep=',', header=0, names=['y', 'datetime'])

# Data preprocessing
# ==============================================================================
data['datetime'] = pd.to_datetime(data['datetime'], format='%Y/%m/%d')
data = data.set_index('datetime')
data = data.asfreq('MS')
data = data['y']
data = data.sort_index()

# Split train-test
# ==============================================================================
steps = 36
data_train = data[:-steps]
data_test  = data[-steps:]

# Plot
# ==============================================================================
fig, ax=plt.subplots(figsize=(9, 4))
data_train.plot(ax=ax, label='train')
data_test.plot(ax=ax, label='test')
ax.legend();
```

### Train forecaster

```python
# Custom function to create predictors
# ==============================================================================
def create_predictors(y):
    """
    Create first 10 lags of a time series.
    Calculate moving average with window 20.
    """

    lags = y[-1:-11:-1]     # window_size = 10
    mean = np.mean(y[-20:]) # window_size = 20
    predictors = np.hstack([lags, mean])

    return predictors
```

```python
# Create and fit forecaster
# ==============================================================================
forecaster = ForecasterAutoregCustom(
                 regressor      = RandomForestRegressor(random_state=123),
                 fun_predictors = create_predictors,
                 name_predictors = [f'lag {i}' for i in range(1, 11)] + ['moving_avg_20'],
                 window_size    = 20 # the window_size needed by the mean is the most restrictive one
             )

forecaster.fit(y=data_train)
forecaster
```

### Prediction

```python
# Predict
# ==============================================================================
steps = 36
predictions = forecaster.predict(steps=steps)
predictions.head(3)
```

```python
# Plot predictions
# ==============================================================================
fig, ax=plt.subplots(figsize=(9, 4))
data_train.plot(ax=ax, label='train')
data_test.plot(ax=ax, label='test')
predictions.plot(ax=ax, label='predictions')
ax.legend();
```

```python
# Prediction error
# ==============================================================================
error_mse = mean_squared_error(
                y_true = data_test,
                y_pred = predictions
            )

print(f"Test error (mse): {error_mse}")
```

### Feature importance

```python
forecaster.get_feature_importance()
```

### Extract training matrix

```python
X, y = forecaster.create_train_X_y(data_train)
X.head()
```

## ForecasterAutoregMultiSeriesCustom

The same existing methods for the `ForecasterAutoregMultiSeries` can be applied to the `ForecasterAutoregMultiSeriesCustom`. Visit [Multi-series forecasting](https://joaquinamatrodrigo.github.io/skforecast/latest/user_guides/multi-time-series-forecasting.html) to learn more about them.

```python
# Data download
# ==============================================================================
url = ('https://raw.githubusercontent.com/JoaquinAmatRodrigo/skforecast/master/' +
       'data/simulated_items_sales.csv')
data = pd.read_csv(url, sep=',')

# Data preparation
# ==============================================================================
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')
data = data.set_index('date')
data = data.asfreq('D')
data = data.sort_index()
data.head()
```

```python
# Split data into train-val-test
# ==============================================================================
end_train = '2014-07-15 23:59:00'
data_train = data.loc[:end_train, :].copy()
data_test  = data.loc[end_train:, :].copy()

print(f"Train dates : {data_train.index.min()} --- {data_train.index.max()}  (n={len(data_train)})")
print(f"Test dates  : {data_test.index.min()} --- {data_test.index.max()}  (n={len(data_test)})")
```

```python
# Plot time series
# ==============================================================================
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 6), sharex=True)

data_train['item_1'].plot(label='train', ax=axes[0])
data_test['item_1'].plot(label='test', ax=axes[0])
axes[0].set_xlabel('')
axes[0].set_ylabel('MW')
axes[0].set_title('Item 1')
axes[0].legend()

data_train['item_2'].plot(label='train', ax=axes[1])
data_test['item_2'].plot(label='test', ax=axes[1])
axes[1].set_xlabel('')
axes[1].set_ylabel('sales')
axes[1].set_title('Item 2')

data_train['item_3'].plot(label='train', ax=axes[2])
data_test['item_3'].plot(label='test', ax=axes[2])
axes[2].set_xlabel('')
axes[2].set_ylabel('sales')
axes[2].set_title('Item 3')

fig.tight_layout()
plt.show();
```

### Custom predictors multi-series

To understand how custom predictors are created in a multi-series forecaster, imagine using the `create_simple_predictors` function with a `pandas DataFrame` of two columns, `item_1` and `item_2`.

The function will generate 3 lags of each time series and the `ForecasterAutoregMultiSeriesCustom` class will join them vertically and also create the columns with the identifiers (`series 1` and `series 2` in the diagram)

<p align="center"><img src="../img/forecaster_multi_series_train_matrix_diagram.png" style="width: 800px"></p>

```python
# Custom function to create predictors
# ==============================================================================
def create_simple_predictors(y):
    """
    Create first 3 lags of a time series.
    """

    lags = y[-1:-4:-1]

    return lags
```

```python
forecaster = ForecasterAutoregMultiSeriesCustom(
                 regressor       = RandomForestRegressor(random_state=123),
                 fun_predictors  = create_simple_predictors,
                 name_predictors = ['lag 1', 'lag 2', 'lag 3'],
                 window_size     = 3 # window needed for 3 lags
             )
```

The final training matrix will be as follows (note that the columns of lags are in a different order than in the diagram but are correctly created).

```python
X = forecaster.create_train_X_y(data_train[['item_1', 'item_2']])[0]
X.head()
```

### Train and predict forecaster

A more complex custom function can help improve the forecaster's predictive capability.

```python
# Custom function to create predictors
# ==============================================================================
def create_complex_predictors(y):
    """
    Create first 10 lags of a time series.
    Calculate moving average with window 20.
    """

    lags = y[-1:-11:-1]     # window_size = 10  
    mean = np.mean(y[-20:]) # window_size = 20
    predictors = np.hstack([lags, mean])

    return predictors
```

```python
# Create and fit forecaster
# ==============================================================================
forecaster = ForecasterAutoregMultiSeriesCustom(
                 regressor       = RandomForestRegressor(random_state=123),
                 fun_predictors  = create_complex_predictors,
                 name_predictors = [f'lag {i}' for i in range(1, 11)] + ['moving_avg_20'],
                 window_size     = 20 # the window_size needed by the mean is the most restrictive one
             )

forecaster.fit(series=data_train)
forecaster
```

```python
# Predict
# ==============================================================================
steps = 36
predictions = forecaster.predict(steps=steps, levels=None)
predictions.head(3)
```

### Extract training matrix

```python
X = forecaster.create_train_X_y(data_train)[0]
X.head()
```

```python
y = forecaster.create_train_X_y(data_train)[1]
y.head()
```
